## Theoretical exercise

1) By definition and differentiation :

$$\hat{\boldsymbol{\beta}}
= \arg\min_{\beta \in \mathbb{R}^{p+1}}
(\mathbf{Y} - \mathbf{X}\beta)^\top \mathbf{W} (\mathbf{Y} - \mathbf{X}\beta)
= (\mathbf{X}^\top \mathbf{W} \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{W}\mathbf{Y}$$

Where $\mathbf{W} = \mathrm{diag}\!\left(
K\!\left(\frac{X_{1}-x}{h}\right), \ldots,
K\!\left(\frac{X_{n}-x}{h}\right)
\right) \in \mathbb{R}^{n \times n}$

So we can choose $w_{ni} = ((\mathbf{X}^\top \mathbf{W} \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{W})_{1,i}$ which not depend on the $Y_i$'s.


2) We compute $\hat{\boldsymbol{\beta}}$ step by step.
$\mathbf{X}^\top \mathbf{W} =
\begin{pmatrix}
K\!\left(\frac{X_{1}-x}{h}\right) & \cdots & K\!\left(\frac{X_{n}-x}{h}\right) \\
(X_1-x)K\!\left(\frac{X_{1}-x}{h}\right) & \cdots & (X_n-x)K\!\left(\frac{X_{n}-x}{h}\right)
\end{pmatrix}$

$\mathbf{X}^\top \mathbf{W} \mathbf{X} = 
\begin{pmatrix}
\sum_i K\!\left(\frac{X_{i}-x}{h}\right) &  \sum_i(X_i-x) K\!\left(\frac{X_{i}-x}{h}\right) \\
\sum_i(X_i-x) K\!\left(\frac{X_{i}-x}{h}\right) & \sum_i(X_i-x)^2 K\!\left(\frac{X_{i}-x}{h}\right) 
\end{pmatrix} = \begin{pmatrix}
nhS_{n,0} &  nhS_{n,1}\\
nhS_{n,1} & nhS_{n,2} 
\end{pmatrix}$

$(\mathbf{X}^\top \mathbf{W} \mathbf{X})^{-1} = \frac{1}{nhS_{n,0}S_{n,2}-nhS_{n,1}^2}\begin{pmatrix}
S_{n,2} &  -S_{n,1}\\
-S_{n,1} & S_{n,0} 
\end{pmatrix}$

$\hat{\boldsymbol{\beta}}_0=((\mathbf{X}^\top \mathbf{W} \mathbf{X})^{-1}\mathbf{X}^\top \mathbf{W}\mathbf{Y})_1 = \frac{1}{nhS_{n,0}S_{n,2}-nhS_{n,1}^2}\sum_j Y_jK\!\left(\frac{X_{j}-x}{h}\right)(S_{n,2} - (X_j-x)S_{n,1})$

So finally $$w_{ni} = \frac{1}{nh}\frac{K\!\left(\frac{X_{i}-x}{h}\right)(S_{n,2} - (X_i-x)S_{n,1})}{S_{n,0}S_{n,2}-S_{n,1}^2}$$  

3) We compute :

$\sum_iw_{ni} = \sum_i \frac{1}{nh}\frac{K\!\left(\frac{X_{i}-x}{h}\right)(S_{n,2} - (X_i-x)S_{n,1})}{S_{n,0}S_{n,2}-S_{n,1}^2} = \frac{1}{nh}\frac{nhS_{n,0}S_{n,2} - nhS_{n,1}^2}{S_{n,0}S_{n,2}-S_{n,1}^2}=1$

## Simulation study: AMISE bandwidth selection

# 1. Aim of the simulation study

The aim of this simulation study is to evaluate the performance of a plug-in estimator
of the AMISE-optimal bandwidth for nonparametric regression with quartic kernel smoothing. 
Specifically, we want to understand:

- how well the block-based estimation of the quantities $\theta_{22}$ and $\sigma^2$ works in practice,

- how the choice of the number of blocks $N$ (selected via Mallows’ $C_p$) impacts the stability of the bandwidth estimator,

- how the estimated bandwidth $\hat{h}_{AMISE}$ depends on the sample size $n$ and the distribution of the covariates $X\sim \text{Beta}(\alpha,\beta)$.

# 2. Quantities in the simulation

Fixed quantities:

- Regression function: 
    $$m(x) = \sin\!\left(\bigl(x/3+0.1\bigr)^{-1}\right).$$
  This function has oscillations and curvature, which makes the estimation of second derivatives $m''$ nontrivial.
- Support of $X$: $[0,1]$, since $X$ is Beta-distributed.
- Simulation parameters: values of $n$, Beta parameters $(\alpha,\beta)$, and variance $\sigma^2=1$.

Random quantities:

- $ X_i \sim \text{Beta}(\alpha,\beta) $, 
- $Y_i = m(X_i) + \varepsilon_i, \quad \varepsilon_i \sim N(0,\sigma^2)$.


Estimated quantities:

- Local quartic polynomial fits $\widehat m_j$ on each block $j=1,\dots,N$,
- Second derivative estimates $\widehat m_j''(x)$,
- $\hat\theta_{22}(N) = \frac{1}{n}\sum_{i=1}^n \big(\widehat m''(X_i)\big)^2,$
- $\hat\sigma^2(N) = \frac{1}{n-5N} \sum_{i=1}^n \big(Y_i - \widehat m(X_i)\big)^2,$
- Plug-in bandwidth:
    $$\hat h_{\text{AMISE}} = n^{-1/5}\!\left(\frac{35\,\hat\sigma^2\,|\text{supp}(X)|}{\hat\theta_{22}}\right)^{1/5},$$
- Cp criterion:
      $$ C_p(N) = \frac{\text{RSS}(N)}{\text{RSS}(N_{\max})/(n-5N_{\max})} - (n-10N).$$
    
# 3. Findings


\subsection*{3. Findings}

The simulation results can be summarized in four main illustrations.  

First, Figure~\ref{fig:N_effect} shows the effect of the number of blocks $N$ on the estimated bandwidth $\hat h$ and on Mallows’ $C_p$. When $N$ is small, each block contains many observations, which produces stable polynomial fits but also biased curvature estimates, leading to bandwidths that are systematically too large. As $N$ increases, the blocks shrink and the quartic fits become unstable, causing large variability in both $\hat\theta_{22}$ and $\hat h$. The $C_p$ criterion successfully identifies an intermediate value of $N$ that balances bias and variance, thereby providing a stable choice of bandwidth.  

Second, Figure~\ref{fig:size_n} illustrates the effect of the sample size $n$. As expected, the distribution of $\hat h$ shifts towards smaller values as $n$ grows, consistent with the theoretical scaling law $h \propto n^{-1/5}$. Moreover, the spread of the boxplots decreases with larger $n$, confirming that the plug-in estimator becomes more stable with more data.  

Third, Figures~\ref{fig:beta_densities} and~\ref{fig:beta_effect} investigate the role of the covariate distribution. Figure~\ref{fig:beta_densities} shows the shapes of several Beta distributions used for the covariates. The Beta(0.5,0.5) distribution produces a U-shape, concentrating data near the boundaries and leaving the middle region sparse. Beta(2,2) yields a bell-shaped distribution with higher density in the center, while Beta(5,1) produces a strongly skewed distribution concentrated near zero. Figure~\ref{fig:beta_effect} then shows how these different covariate distributions affect the bandwidth estimator. The results clearly demonstrate that homogeneous covariate distributions such as Beta(2,2) lead to more stable estimates of $\hat h$. In contrast, unbalanced or U-shaped distributions increase the variability of $\hat h$, since sparse regions make the estimation of second derivatives less reliable.  

Overall, these figures confirm that (i) the block size $N$ must be chosen carefully, with Mallows’ $C_p$ providing a suitable selection rule, (ii) the sample size $n$ drives both the rate and stability of the estimator in agreement with asymptotic theory, and (iii) the shape of the covariate distribution strongly affects the variability of the plug-in estimator, with dense and balanced covariate distributions being most favorable.
